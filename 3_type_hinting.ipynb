## Type Hints and Annotation, PEP 484, Python 3.5+

### Benefits

- Trade off between extra work and readability
- Better understand of your code
- Help IDE to catch errors


```python
# No Type hinting
def device_upper(name):
    return name.upper()
```

```python
print(device_upper('lax-edg-r1'))
```

```python
print(device_upper(10))
```

```python
# With Type hinting
def device_upper_2(name: str) -> str:
    return name.upper()
```

```python
# not actually enforced
print(device_upper_2(10))
```
