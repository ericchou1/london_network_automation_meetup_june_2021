## Type Hints and Annotation, PEP 484, Python 3.5+

### Benefits

- More readiable code for your future self and your team (auto documentation? maybe)
- Trade off between extra work and readability
- Better understand of your code
- Help IDE to catch errors


```python
# No Type hinting
# What is the type of argument and what is the return value type?  
# What if this is imported or this was not on our screen? 
def device_upper(name):
    return name.upper()
```

```python
print(device_upper('lax-edg-r1'))
```

```python
print(device_upper(10))
```

```python
# With Type hinting, argument data type and return value are specified
def device_upper_2(name: str) -> str:
    return name.upper()
```

```python
# not actually enforced
print(device_upper_2(10))
```
